```python
# Import data
# df2 = pd.read_csv('henrick_mock_data_table.csv', sep=",", index_col="Contig", na_values=['-'])
df2 = pd.read_csv('henrick_revised_timeline.csv', sep=",", na_values=['-'])
```

```python
sns.set(style="whitegrid")

# Initialize the figure
f, ax = plt.subplots(figsize=(16,8))
# sns.despine(bottom=True, left=True)

# Show each observation with a scatterplot
sns.stripplot(x="Year", y="Cases", hue="Virus",
              data=df2, palette=col_list_palette, marker="o", s=25, jitter= True,edgecolor="black",linewidth=2, size=15)


# sns.stripplot(x="Year", y="Cases", hue="Virus",
#               data=df2, palette=col_list_palette, marker="o", s=16, alpha =0.8)

# x_ticks =['1955','1976','1990','1995','2010','2019']
# ax.set_xticklabels(df2.Year, size = 16)
# ax.set_yticklabels(size = 16)
# ax.set_xticklabels(x_ticks, rotation=90, size = 16)

# # ticks_hh=[] 
# # for i in range(len(virus_list)+1):
# #     ticks_hh.append(i)
# # ax.set_yticks(ticks_hh)
# ax.set_yticklabels(virus_list, size = 16)
# ax.set_yticklabels(virus_list, rotation=0)
ax.grid(which='major', linestyle=':', linewidth='0.5', color='grey')
# ax.patch.set_facecolor('#cccccc')#a grey
ax.patch.set_facecolor('#FFFFFF')#white
# plt.title("VHF Outbreak cases timeline", size = 20)
plt.xticks(size=15)
plt.yticks(size=15)
plt.xlabel("Outbreak year", size=15, fontweight="bold")
plt.ylabel("Number of cases", size=15,fontweight="bold")
plt.savefig('Henrick_VHF_timeline.pdf',bbox_inches="tight")
plt.savefig('Henrick_VHF_timeline.jpg',bbox_inches="tight", dpi=300)
```
